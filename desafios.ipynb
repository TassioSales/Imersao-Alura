### Solução desafio 01 segundo dia

```python
mapa = {'droga':'composto'}
dados.rename(columns = mapa, inplace = True)
dados.head()

top5_compostos = dados['composto'].value_counts().index[:5]
top5_compostos


import seaborn as sns
import matplotlib.pyplot as plt

sns.set_palette('Accent')
sns.set()
plt.figure(figsize = (12, 8))
ax = sns.countplot(x = 'composto',data = dados.query('composto in @top5_compostos'), 
                   order = dados['composto'].value_counts().head().index)
plt.show()
```

## Solução desafio 02 segundo dia

```python
sns.set_palette('Accent')
sns.set()
plt.figure(figsize = (12, 8))
ax = sns.countplot(x = 'composto',data = dados.query('composto in @top5_compostos'), 
                   order = dados['composto'].value_counts().head().index, palette="flare")
ax.set_title('Top 5 composto', loc ='center', fontsize = 18)
ax.set_xlabel('composto', fontsize = 14)
ax.set_ylabel('Nome dos compostos', fontsize = 14)
plt.show()
```
